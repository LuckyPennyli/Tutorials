```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

random = np.random.rand(100, 2)
plt.figure(figsize=(12, 3))
for k in range(1,6):
    kmeans = KMeans(n_clusters=k)
    a = kmeans.fit_predict(random)
    plt.subplot(1,5,k)
    plt.scatter(random[:, 0], random[:, 1], c=a)
    plt.xlabel('k='+str(k))
plt.tight_layout()
plt.show()
```
