```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
train_identity = pd.read_csv("train_identity.csv")
train_transaction = pd.read_csv("train_transaction.csv")
df_train=pd.merge(train_transaction, train_identity, on="TransactionID", how="left")
df_train
```

```python
for col in df_train:
    print(col)
```

```python
filter_col = [col for col in df_train if col.startswith('V')]
df_filtered = df_train[filter_col]
df_filtered
```

```python
filter_col_first_150 = df_filtered.iloc[: , :150]
filter_col_first_150
```

```python
filter_col_first_150.corr()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(filter_col_first_150.corr())
plt.show()
```
