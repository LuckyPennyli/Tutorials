```python
import pandas as pd
import numpy as np
X = pd.read_csv('CleansedWhiteCollar.csv')
```

```python
table = pd.pivot_table(X, values=['AMOUNT'], index=['Vendor Number'],columns=['Department'], aggfunc=np.sum)
table
```

```python
table = pd.pivot_table(X, values=['AMOUNT'], index=['Vendor Number'],columns=['Department'], aggfunc=np.sum, fill_value=0)
table
```

```python
table = pd.pivot_table(X, values = ['AMOUNT'], index =['Vendor Number'], aggfunc={'AMOUNT': np.mean})
table
```

```python
table = pd.pivot_table(X, values = ['AMOUNT'], index=['Vendor Number'], aggfunc={'AMOUNT':[min, max, np.mean]})
table
```
